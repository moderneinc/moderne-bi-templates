# Dashboard KPIs

Executive-level snapshot — KPI cards with a monthly trend chart.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Load sample data — replace with your own query results
summary = pd.read_csv("../../samples/dashboard-kpis-summary.csv")
trend = pd.read_csv("../../samples/dashboard-kpis-trend.csv", parse_dates=["month"])
```

```python
# KPI cards
kpis = [
    ("Total Recipe Runs", f"{summary['total_recipe_runs'].iloc[0]:,}"),
    ("Total Commits", f"{summary['total_commits'].iloc[0]:,}"),
    ("Unique Users", f"{summary['unique_users'].iloc[0]:,}"),
    ("Unique Recipes", f"{summary['unique_recipes'].iloc[0]:,}"),
    ("Files Changed", f"{summary['files_changed'].iloc[0]:,}"),
    ("Est. Hours Saved", f"{summary['estimated_hours_saved'].iloc[0]:,.1f}"),
]

colors = ["#283AF7", "#7C3AED", "#E93D82", "#283AF7", "#7C3AED", "#E93D82"]

fig, axes = plt.subplots(1, len(kpis), figsize=(18, 3))
for ax, (label, value), color in zip(axes, kpis, colors):
    ax.text(0.5, 0.55, value, ha="center", va="center", fontsize=28,
            fontweight="bold", color=color, transform=ax.transAxes)
    ax.text(0.5, 0.15, label, ha="center", va="center", fontsize=11,
            color="black", transform=ax.transAxes)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis("off")

fig.suptitle("Dashboard KPIs", fontsize=16, fontweight="bold", color="black", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Monthly trend
fig, ax1 = plt.subplots(figsize=(14, 7))

bar_width = 8

ax1.bar(trend["month"] - pd.Timedelta(days=bar_width / 2), trend["recipe_runs"],
        width=bar_width, color="#283AF7", alpha=0.8, label="Recipe Runs")
ax1.bar(trend["month"] + pd.Timedelta(days=bar_width / 2), trend["commits"],
        width=bar_width, color="#7C3AED", alpha=0.8, label="Commits")

ax1.set_xlabel("Month", color="black")
ax1.set_ylabel("Count", color="black")
ax1.tick_params(axis="both", labelcolor="black")

# Secondary y-axis for estimated hours saved
ax2 = ax1.twinx()
ax2.plot(trend["month"], trend["estimated_hours_saved"], color="#E93D82", marker="D",
         linewidth=2, linestyle="-", label="Est. Hours Saved")
ax2.set_ylabel("Estimated Hours Saved", color="black")
ax2.tick_params(axis="y", labelcolor="black")

ax1.set_title("Monthly Trend", color="black")
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
ax1.xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45, ha="right")

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left", fontsize=8)

plt.tight_layout()
plt.show()
```
