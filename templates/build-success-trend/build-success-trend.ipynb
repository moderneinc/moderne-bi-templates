# Build Success Trend

Monthly build health — successful vs. failed builds with success rate overlay.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Load sample data — replace with your own query results
df = pd.read_csv("../../samples/build-success-trend.csv", parse_dates=["month"])
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

bar_width = 12

# Stacked bars: successful on bottom, failed on top
ax1.bar(df["month"], df["successful_builds"],
        width=bar_width, color="#283AF7", alpha=0.8, label="Successful Builds")
ax1.bar(df["month"], df["failed_builds"], bottom=df["successful_builds"],
        width=bar_width, color="#E93D82", alpha=0.8, label="Failed Builds")

ax1.set_xlabel("Month", color="black")
ax1.set_ylabel("Builds", color="black")
ax1.tick_params(axis="both", labelcolor="black")

# Secondary y-axis for success rate
ax2 = ax1.twinx()
ax2.plot(df["month"], df["success_rate_pct"], color="#7C3AED", marker="D",
         linewidth=2, linestyle="-", label="Success Rate %")
ax2.set_ylabel("Success Rate %", color="black")
ax2.set_ylim(80, 100)
ax2.tick_params(axis="y", labelcolor="black")

ax1.set_title("Build Success Trend", color="black")
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
ax1.xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45, ha="right")

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left", fontsize=8)

plt.tight_layout()
plt.show()
```
