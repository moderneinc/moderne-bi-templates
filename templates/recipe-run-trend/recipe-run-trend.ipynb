# Recipe Run Trend

Monthly adoption trend showing recipe runs, distinct recipes, and unique users over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Load sample data — replace with your own query results
df = pd.read_csv("../../samples/recipe-run-trend.csv", parse_dates=["month"])
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

bar_width = 8
offsets = [-bar_width, 0, bar_width]
colors = ["#283AF7", "#7C3AED", "#E93D82"]
labels = ["Recipe Runs", "Distinct Recipes", "Unique Users"]
columns = ["recipe_runs", "distinct_recipes", "unique_users"]

for offset, color, label, col in zip(offsets, colors, labels, columns):
    ax.bar(
        df["month"] + pd.Timedelta(days=offset),
        df[col],
        width=bar_width,
        color=color,
        label=label,
    )

ax.set_xlabel("Month")
ax.set_ylabel("Count")
ax.set_title("Recipe Run Trend")
ax.legend()
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
