# Top Users

User engagement ranking by recipe runs and commits.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load sample data — replace with your own query results
df = pd.read_csv("../../samples/top-users.csv")
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

bar_height = 0.35
y = range(len(df))

ax.barh([i + bar_height / 2 for i in y], df["recipe_runs"], bar_height, color="#283AF7", label="Recipe Runs")
ax.barh([i - bar_height / 2 for i in y], df["commits"], bar_height, color="#E93D82", label="Commits")

ax.set_yticks(list(y))
ax.set_yticklabels(df["developer"])
ax.invert_yaxis()
ax.set_xlabel("Count")
ax.set_title("Top Users")
ax.legend()
plt.tight_layout()
plt.show()
```
