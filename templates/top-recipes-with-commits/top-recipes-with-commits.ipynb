# Top Recipes with Commits

Recipes ranked by committed code changes — distinguishes recipes that were tried from recipes that delivered results.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load sample data — replace with your own query results
df = pd.read_csv("../../samples/top-recipes-with-commits.csv")
```

```python
fig, ax1 = plt.subplots(figsize=(12, 7))

bar_height = 0.35
y = range(len(df))

ax1.barh([i - bar_height / 2 for i in y], df["commits"], bar_height, color="#283AF7", label="Commits")
ax1.barh([i + bar_height / 2 for i in y], df["unique_users"], bar_height, color="#E93D82", label="Unique Users")

ax1.set_yticks(list(y))
ax1.set_yticklabels(df["recipe_name"])
ax1.invert_yaxis()
ax1.set_xlabel("Count")
ax1.set_xlim(0, 500)
ax1.set_title("Top Recipes with Commits")

# Overlay estimated hours saved as text annotations
for i, hours in enumerate(df["estimated_hours_saved"]):
    ax1.text(df["commits"].iloc[i] + 5, i - bar_height / 2, f"{hours}h",
             va="center", fontsize=8, color="#7C3AED")

ax1.legend()
plt.tight_layout()
plt.show()
```
