# Top Recipes

Most-used recipes ranked by run count, unique users, and repos searched.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load sample data — replace with your own query results
df = pd.read_csv("../../samples/top-recipes.csv")
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

bar_height = 0.25
y = range(len(df))

ax.barh([i - bar_height for i in y], df["repos_searched"], bar_height, color="#283AF7", label="Repos Searched")
ax.barh(list(y), df["unique_users"], bar_height, color="#7C3AED", label="Unique Users")
ax.barh([i + bar_height for i in y], df["recipe_runs"], bar_height, color="#E93D82", label="Recipe Runs")

ax.set_yticks(list(y))
ax.set_yticklabels(df["recipe_name"])
ax.invert_yaxis()
ax.set_xlabel("Count")
ax.set_title("Top Recipes")
ax.legend()
plt.tight_layout()
plt.show()
```
