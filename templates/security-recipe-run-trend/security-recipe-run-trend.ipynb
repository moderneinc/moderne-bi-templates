# Security Recipe Run Trend

Monthly trend of committed security remediation — fixes that actually landed in code.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Load sample data — replace with your own query results
df = pd.read_csv("../../samples/security-recipe-run-trend.csv", parse_dates=["month"])
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

bar_width = 6

# Bars for volume metrics
ax1.bar(df["month"] - pd.Timedelta(days=bar_width), df["successful_commits"],
        width=bar_width, color="#283AF7", alpha=0.8, label="Successful Commits")
ax1.bar(df["month"], df["unique_repos_fixed"],
        width=bar_width, color="#7C3AED", alpha=0.8, label="Unique Repos Fixed")
ax1.bar(df["month"] + pd.Timedelta(days=bar_width), df["distinct_recipes"],
        width=bar_width, color="#E93D82", alpha=0.8, label="Distinct Recipes")

ax1.set_xlabel("Month", color="black")
ax1.set_ylabel("Count", color="black")
ax1.tick_params(axis="both", labelcolor="black")

# Line overlay for files remediated
ax1.plot(df["month"], df["files_remediated"], color="#00CFF3", marker="o",
         linewidth=2, linestyle="--", label="Files Remediated")

# Secondary y-axis for estimated hours saved
ax2 = ax1.twinx()
ax2.plot(df["month"], df["estimated_hours_saved"], color="#85FE99", marker="D",
         linewidth=2, linestyle="-", label="Est. Hours Saved")
ax2.set_ylabel("Estimated Hours Saved", color="black")
ax2.tick_params(axis="y", labelcolor="black")

ax1.set_title("Security Recipe Run Trend", color="black")
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
ax1.xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45, ha="right")

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left", fontsize=8)

plt.tight_layout()
plt.show()
```
